```python
from sklearn.preprocessing import StandardScaler
def get_data(ticker_list, start, end):
    return_list = []
    for i in ticker_list:
        df = pd.DataFrame(get_historical_data(i, start, end, output_format='pandas')).interpolate()
        df['ticker'] = i
        return_list.append(df)
    return return_list

def lstm_clean_data(dataset):
    for i in range(len(dataset)):
        dataset[i]['Reg_Target'] = dataset[i]['close'].shift(-30)
#         dataset[i]['Reg_Target']=(np.sign(-dataset[i]['close'].diff(periods=-1))+1)/2
#         dataset[i]['Reg_Target'] = dataset[i]['close'].shift(-1)                  #price difference from the prevous date
#         dataset[i]['Reg_Target']=(np.sign(-data[1]['close'].diff(periods=-15))+1)/2
#         dataset[i]['Reg_Target'] = -dataset[i]['close'].diff(periods=-1)
        
        dataset[i]['f1'] = dataset[i]['volume'].apply(np.log)
        dataset[i]['f2']= dataset[i]['volume'].diff()
        dataset[i]['f3']= dataset[i]['volume'].diff(10)
        #percent change 
        dataset[i]['f05'] = (dataset[i]['volume']-dataset[i]['volume'].shift(-1))/dataset[i]['volume'] #perfect change in volume
        dataset[i]['f06'] = (dataset[i]['close']-dataset[i]['close'].shift(-1))/dataset[i]['close'] #percent change in close 
        dataset[i]['f07'] = dataset[i]['close']/(dataset[i]['close'].shift(1).rolling(window = 30).mean())
        #high Low 
        dataset[i]['f8'] = dataset[i]['high'] - dataset[i]['low']
        dataset[i]['f9'] = dataset[i]['close'] - dataset[i]['open']
        #3day MA'
        dataset[i]['f10'] = dataset[i]['close'].shift(1).rolling(window = 3).mean()
        dataset[i]['f11'] = np.log(dataset[i]['close'].shift(1).rolling(window = 3).mean())
        dataset[i]['f13'] = np.log(dataset[i]['close'].shift(1).rolling(window = 10).mean())

        #30 day MA 
        dataset[i]['f14'] = dataset[i]['close'].shift(1).rolling(window = 30).mean()
        dataset[i]['f15'] = np.log(dataset[i]['close'].shift(1).rolling(window = 30).mean())

        #STD
        dataset[i]['f16']= dataset[i]['close'].rolling(5).std()
        dataset[i]['f17']= dataset[i]['close'].rolling(10).std()
        dataset[i]['f18']= dataset[i]['close'].rolling(30).std()
        dataset[i]['f19']= (dataset[i]['close']-dataset[i]['f14'])/dataset[i]['f18'] #kind of Z-score using 30 day std
        dataset[i]['f19']= (dataset[i]['close']-dataset[i]['f14'])/dataset[i]['f17'] #kind of Z-score using 10 day std

        #10day MA
        dataset[i]['f12'] = dataset[i]['close'].shift(1).rolling(window = 10).mean()
        #reset and drop
        dataset[i] = dataset[i].reset_index().dropna()
        dataset[i]['date'] = pd.to_datetime(dataset[i]['date'])
        dataset[i] = dataset[i].set_index('date')
        
    return dataset

def add_past(etf_list, times):
    for i in range(len(etf_list)):
        for n in times:
            etf_list[i]['{}day return'.format(n)] = -etf_list[i]['close'].diff(periods=n).round(2)
            etf_list[i]=etf_list[i].dropna()
    return etf_list

def lstm_test_split(X, n_past, date):
    X = X.reset_index()
    scaler = StandardScaler()
    y_scaler = StandardScaler()
    ticker = X['ticker'].iloc[0]
    x_train = X[X['date'] < date].drop(columns=['date', 'Reg_Target', 'ticker','volume'])
    scaler.fit(x_train)
    x_test = X[X['date'] >= date].drop(columns=['date', 'Reg_Target', 'ticker','volume'])[:-1] #, '1day return', '5day return', '21day return', '35day return'
    
    x_train = scaler.transform(x_train)
    x_train = np.reshape(x_train,(x_train.shape[0], n_past, x_train.shape[1]))
    x_test = scaler.transform(x_test)
    x_test = np.reshape(x_test,(x_test.shape[0], n_past, x_test.shape[1]))
    
    y_train = np.array(X[X['date'] < date]['Reg_Target'].drop(columns='date')).ravel().astype('float').reshape(-1,1)
    y_scaler.fit(y_train)
    y_train = y_scaler.transform(y_train)
    
    y_test = np.array(X[X['date'] >= date]['Reg_Target'].drop(columns='date')).ravel().astype('float')[:-1].reshape(-1,1)
    y_test = y_scaler.transform(y_test)
    x_holdout = X[X['date'] >= date].drop(columns=['date', 'Reg_Target', 'ticker','volume'])[-1:] #'1day return', '5day return', '21day return', '35day return'
    x_holdout = scaler.transform(x_holdout)
    x_holdout = np.reshape(x_holdout,(x_holdout.shape[0], n_past, x_holdout.shape[1]))
#     y_test = scaler.transform(y_test)
    return ticker, x_train, x_test, x_holdout, y_train, y_test, scaler, y_scaler

def build_step_model(x_train, y_train, epoc):    
    model = Sequential()
    model.add(LSTM(50, input_shape=(x_train.shape[1], x_train.shape[2]), return_sequences=True))
    model.add(Dropout(0.2))
    model.add(LSTM(100, return_sequences=False))
    model.add(Dropout(0.2))
    model.add(Dense(1))
    model.add(Activation('relu'))
    model.compile(loss='mse', optimizer='adam')
    history = model.fit(x_train, y_train, epochs=epoc, batch_size=64, validation_split=.1, verbose=2,shuffle=False)
    return model, history

def yield_preds(model, scaler, x_test, x_holdout, y_test):
    yhat = model.predict(x_test) #dynamic=False
    preds = scaler.inverse_transform(yhat)
    true = scaler.inverse_transform(y_test)
    today = model.predict(x_holdout)
    today_pred = scaler.inverse_transform(today)
    return preds, today_pred

def run_all_lstms(data, split, epoc):
    out = pd.DataFrame()
    tomorrow = pd.DataFrame()
    start = timeit.default_timer()
    for i in range(len(data)):
        ticker, x_train, x_test, x_holdout, y_train, y_test, scaler, y_scaler = lstm_test_split(data[i], 1, split)
        print('Model #: {}'.format(i))
        model, history = build_step_model(x_train, y_train, epoc)
        preds, future = yield_preds(model, y_scaler, x_test, x_holdout, y_test)
        out[ticker] = preds.flatten()
        tomorrow[ticker] = future.flatten()
    out = out.set_index(data[0][-len(out):].index)
    stop = timeit.default_timer()
    print('Time: ', stop - start)
    return out, tomorrow


etf_list2 =ticker_list[0:9] #
etf_list1 =ticker_list[9:19]
etf_list3 =ticker_list[19:] #28 tickers 
data1 = get_data(etf_list1, None, None)  #max call on get_iex have to break up
# @changed batchsie to smaller from 64 to 35 
```

```python
from sklearn.metrics import roc_curve, auc
def run_metrics(y_true, y_pred, ticker):
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred,average=None)
    f1 = f1_score(y_true, y_pred,average=None)
    roc = f1_score(y_true, y_pred,average=None)       #roc_auc_score(y_true, y_pred, average="weighted")
    return {'!Ticker': ticker, 'Accuracy': accuracy, 'Precision': precision, 'F1': f1, 'ROC AUC': roc}

def run_random_forest(df_list):
    results = []
    for i in range(len(df_list)):
        ticker = df_list[i]['ticker'].iloc[0]
        df_list[i][15:-30].dropna()
        X=df_list[i][15:-30].drop(columns=['Date','Reg_Target','ticker',"volume"])
    #     y=df_list[i][15:-10]['Prediction'].ravel().astype('int')
        y=np.array(df_list[i][15:-30]['Prediction']).ravel().astype('int')
    # results = []
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .9)

        clf = RandomForestClassifier(n_estimators=10, max_depth=4)
        clf.fit(X_train, y_train)
        y_pred = clf.predict(X_test)
        results.append(run_metrics(y_test, y_pred, ticker))       
```
